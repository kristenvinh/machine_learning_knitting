```python
#Read in the attributes data
df_attributes = pd.read_csv('../data_aquisition/ravelry_pattern_attributes.csv')
```

```python
# Get all possible attributes
attributes = df_attributes['attribute_name'].to_list()
```

```python
for attribute in attributes:
    df[attribute] = df['Attributes'].apply(lambda attribute_list: attribute in attribute_list)
```

```python
# Use .isin() to check for multiple values at once
df = df[df['Pattern Type'].isin(['pullover', 'cardigan'])]
```

```python
# Isolate only the attribute columns (i.e., not the ID columns)
attribute_only_cols = df[attributes]

# Sum() on boolean columns counts the 'True' values (True=1, False=0)
attribute_counts = attribute_only_cols.sum().reset_index()

# Rename the columns for clarity
attribute_counts.columns = ['Attribute', 'Count']

# Sort the counts from most common to least common
attribute_counts = attribute_counts.sort_values(by='Count', ascending=False)
```

```python
#Join Attribute Counts with Attribute Descriptions for attribute decision making
df_attributes = df_attributes.join(attribute_counts.set_index('Attribute'), on='attribute_name')
```

```python
df_attributes.to_csv('attribute_counts_with_descriptions.csv', index=False)
```

```python
# Define the columns that you want to use for filtering.
filter_cols = ['baby',
 'child (4-12)',
 'doll',
 'newborn',
 'preemie',
 'teen (13-17)',
 'toddler (1-3)',]

# We need to find which of these columns actually exist in our dataframe,
# in case some attributes were not present in the sample data.
existing_filter_cols = [col for col in filter_cols if col in df.columns]

# Remove rows where ANY of the specified columns have a True value.
# The `~` symbol negates the condition, so we keep rows where the condition is False.
filtered_df = df[~df[existing_filter_cols].any(axis=1)]
```

```python
filtered_df = filtered_df.drop(columns=existing_filter_cols)
```

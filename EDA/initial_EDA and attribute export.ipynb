```python
import os
import glob
import pandas as pd

# Define the paths to both directories
output_dir1 = "../data_aquisition/processed_chunks"
output_dir2 = "../data_aquisition/processed_chunks2"

# Find all the CSV files in the first directory
chunk_files1 = glob.glob(os.path.join(output_dir1, "*.csv"))

# Find all the CSV files in the second directory
chunk_files2 = glob.glob(os.path.join(output_dir2, "*.csv"))

# Combine the two lists of file paths
all_chunk_files = chunk_files1 + chunk_files2

# Read each chunk file into a list of dataframes
# This part remains the same, just using the combined list
df_list = [pd.read_csv(f) for f in all_chunk_files]

# Concatenate all dataframes in the list into one
df = pd.concat(df_list, ignore_index=True)
```

```python
# Use .isin() to check for multiple values at once
# Get only pullovers and cardigans
df = df[df['Pattern Type'].isin(['pullover', 'cardigan'])]
```

```python
df['Attributes_list'] = df['Attributes'].fillna('[]').apply(ast.literal_eval)

# Create a new DataFrame with dummy variables from the 'Attributes_list'
# We first explode the lists to have one attribute per row, then get dummies
# and finally group by the original index and sum to get the one-hot encoding
attributes_dummies = df['Attributes_list'].explode().str.get_dummies().groupby(level=0).sum()

# Add a prefix to the attribute columns to avoid potential column name conflicts
attributes_dummies = attributes_dummies.add_prefix('attr_')

# Concatenate the new dummy variables DataFrame with the original DataFrame
df_expanded = pd.concat([df, attributes_dummies], axis=1)

# Drop the intermediate 'Attributes_list' and original 'Attributes' columns
df_expanded = df_expanded.drop(['Attributes', 'Attributes_list'], axis=1)
```

```python
# Now, filter the DataFrame to include only rows where 'attr_adult' is True
df_expanded = df_expanded[df_expanded['attr_adult'] == True]

df_expanded.shape
```

```python
   # Get all the columns that start with 'attr_'
attribute_columns = [col for col in df_expanded.columns if col.startswith('attr_')]

# Calculate the sum of each attribute column
attribute_counts = df_expanded[attribute_columns].sum()

# Create a new DataFrame from the attribute counts
attribute_counts_df = pd.DataFrame(attribute_counts, columns=['Count'])

# Reset the index to make the attribute names a column
attribute_counts_df = attribute_counts_df.reset_index().rename(columns={'index': 'Attribute'})

# Sort the attributes by count in descending order
attribute_counts_df = attribute_counts_df.sort_values(by='Count', ascending=False)

# Remove the 'attr_' prefix from the 'Attribute' column
attribute_counts_df['Attribute'] = attribute_counts_df['Attribute'].str.replace('attr_', '')
attribute_counts_df.to_csv('attribute_counts.csv', index=False)
```
